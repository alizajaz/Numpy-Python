```python
#draw two line
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(y,marker='o',ms=20, mec='hotpink', mfc='hotpink', linestyle='dotted', color='g',linewidth='10')
plt.plot(y,marker='*',ms=20, mec='g', mfc='g', linestyle='dotted', color='r', linewidth='10')
plt.show()
```

```python
#draw two line
#title
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(y,marker='o',ms=20, mec='hotpink', mfc='hotpink', linestyle='dotted', color='g',linewidth='10')
plt.plot(y,marker='*',ms=20, mec='g', mfc='g', linestyle='dotted', color='r', linewidth='10')
plt.title("X vs Y")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
#grid or boxes on background
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(y,marker='o',ms=20, mec='hotpink', mfc='hotpink', linestyle='dotted', color='g',linewidth='10')
plt.plot(y,marker='*',ms=20, mec='g', mfc='g', linestyle='dotted', color='r', linewidth='10')
plt.title("X vs Y")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

```python
#draw multiple graph on same cell
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([0,1,2,3])
y=np.array([3,18,1,10])
plt.subplot(1,2,1)#the graph has 1 row, 2 columns graph no 1 
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([10,18,21,30])
plt.subplot(1,2,2)
plt.plot(x,y)


plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np 
x=np.array([0,1,2,3])
y=np.array([3,18,1,10])
plt.subplot(2,2,1)#the graph has 2 row, 1 columns graph no 1 
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([10,18,21,30])
plt.subplot(2,1,2)#the graph has 2 row, 1 columns graph no 1 
plt.plot(x,y)
plt.title("Income")


plt.show()
```
